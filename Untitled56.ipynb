```python
# %% Install (first run only)
try:
    get_ipython
    %pip -q install pandas numpy scikit-learn matplotlib seaborn joblib beautifulsoup4 html5lib ipywidgets
except Exception:
    pass

# %% Imports & paths
from pathlib import Path
import re, string
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import joblib

# Your file + output dir (edit if needed)
DATA = Path(r"C:\Users\NXTWAVE\Downloads\Review Sentiment Analyzer\archive\Reviews.csv")
OUT_DIR = Path(r"C:\Users\NXTWAVE\Downloads\Review Sentiment Analyzer")
OUT_DIR.mkdir(parents=True, exist_ok=True)

# Candidate columns
TEXT_CANDS = ["reviewText","Text","text","review","content"]
RATING_CANDS = ["Score","Rating","rating","stars"]
LABEL_CANDS = ["label","sentiment","polarity"]

# --- cleaning helpers ---
URL_RE = re.compile(r"https?://\S+|www\.\S+", re.IGNORECASE)
PUNCT_TABLE = str.maketrans('', '', string.punctuation)

def clean_text(s: str) -> str:
    if not isinstance(s, str): return ""
    t = BeautifulSoup(s, "html5lib").get_text(" ")
    t = URL_RE.sub(" ", t).lower().translate(PUNCT_TABLE)
    t = re.sub(r"[^a-z0-9\s]", " ", t)
    t = re.sub(r"\s+", " ", t).strip()
    return t

def pick_col(df, cands):
    for c in cands:
        if c in df.columns: return c
    return None

# %% Load & prepare labels
df = pd.read_csv(DATA)
text_col = pick_col(df, TEXT_CANDS)
if not text_col:
    raise SystemExit(f"No text column found in {df.columns.tolist()}")

label_col = pick_col(df, LABEL_CANDS)
rating_col = pick_col(df, RATING_CANDS)

df["text"] = df[text_col].astype(str)
df["text_clean"] = df["text"].map(clean_text)

if label_col:
    # normalize labels to 3 classes if present
    lab = df[label_col].astype(str).str.lower().str.strip()
    lab = lab.replace({"pos":"positive","neg":"negative","neu":"neutral",
                       "positive":"positive","negative":"negative","neutral":"neutral"})
    df["label"] = lab
elif rating_col:
    r = pd.to_numeric(df[rating_col], errors="coerce")
    df["label"] = r.apply(lambda x: "positive" if x>=4 else ("negative" if x<=2 else "neutral"))
else:
    raise SystemExit("No label/sentiment/Rating column found. Add one or map your columns.")

# Drop rows missing labels or text
df = df.dropna(subset=["text_clean","label"]).copy()

# Optionally, if neutral is tiny, drop it for a cleaner binary model
if (df["label"]=="neutral").sum() < 50:
    df = df[df["label"]!="neutral"].copy()

print("Label counts:\n", df["label"].value_counts())

# %% Train/valid split + model
Xtr, Xva, ytr, yva = train_test_split(
    df["text_clean"], df["label"], test_size=0.2, random_state=42, stratify=df["label"]
)

pipe = Pipeline([
    ("tfidf", TfidfVectorizer(max_features=50000, ngram_range=(1,2))),
    ("clf", LogisticRegression(C=4.0, max_iter=250, class_weight="balanced", n_jobs=-1))
])
pipe.fit(Xtr, ytr)

# Save model (optional)
joblib.dump(pipe, OUT_DIR / "model_tfidf_lr.joblib")

# %% Evaluate
yhat = pipe.predict(Xva)
report = classification_report(yva, yhat, output_dict=True)
labels_sorted = sorted(df["label"].unique().tolist())
cm = confusion_matrix(yva, yhat, labels=labels_sorted)

print(pd.DataFrame(report).round(3))
print("\nConfusion matrix:\n", cm)

# %% Accuracy graph (overall accuracy + macro/weighted F1)
acc = float(report.get("accuracy", 0.0))
macro_f1 = float(report.get("macro avg", {}).get("f1-score", 0.0))
weighted_f1 = float(report.get("weighted avg", {}).get("f1-score", 0.0))

plt.figure()
ax = sns.barplot(x=["accuracy","macro_f1","weighted_f1"], y=[acc, macro_f1, weighted_f1])
ax.set_ylim(0, 1); ax.set_title("Model accuracy & F1")
for c in ax.containers: ax.bar_label(c, fmt="%.3f")
plt.tight_layout()
plt.savefig(OUT_DIR / "viz_accuracy.png", dpi=160)
plt.show()

# %% Per-class Precision/Recall/F1 bars
class_rows = {k:v for k,v in report.items() if isinstance(v, dict) and k not in ("accuracy","macro avg","weighted avg")}
if class_rows:
    labels = list(class_rows.keys())
    prec  = [class_rows[k].get("precision",0) for k in labels]
    reca  = [class_rows[k].get("recall",0) for k in labels]
    f1    = [class_rows[k].get("f1-score",0) for k in labels]
    x = np.arange(len(labels)); w=0.25
    plt.figure(figsize=(8,4))
    plt.bar(x-w, prec, width=w, label="precision")
    plt.bar(x,    reca, width=w, label="recall")
    plt.bar(x+w,  f1,   width=w, label="f1")
    plt.xticks(x, labels); plt.ylim(0,1); plt.legend()
    plt.title("Per-class PRF1")
    plt.tight_layout()
    plt.savefig(OUT_DIR / "viz_prf1_by_class.png", dpi=160)
    plt.show()

# %% Confusion matrix heatmaps (counts + row-normalized)
plt.figure()
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=labels_sorted, yticklabels=labels_sorted)
plt.xlabel("Predicted"); plt.ylabel("True"); plt.title("Confusion Matrix (counts)")
plt.tight_layout(); plt.savefig(OUT_DIR / "viz_confusion_matrix.png", dpi=160); plt.show()

row_sums = cm.sum(axis=1, keepdims=True)
norm = cm / np.where(row_sums==0, 1, row_sums)
plt.figure()
sns.heatmap(norm, annot=True, fmt=".2f", cmap="Greens", xticklabels=labels_sorted, yticklabels=labels_sorted)
plt.xlabel("Predicted"); plt.ylabel("True"); plt.title("Confusion Matrix (row-normalized)")
plt.tight_layout(); plt.savefig(OUT_DIR / "viz_confusion_matrix_normalized.png", dpi=160); plt.show()

print("Saved plots to:", OUT_DIR)
```
